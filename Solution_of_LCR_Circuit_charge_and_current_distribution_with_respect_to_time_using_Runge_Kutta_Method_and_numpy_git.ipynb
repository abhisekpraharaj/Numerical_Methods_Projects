Case_3 Under Damped

```python
"""initial value of t,i
###############################################Case-3, Underdamped case R>2*Sqrt(L/C)
e=1.0
l=1.0
c=0.5
r=0.5
##############################################################"""
```

```python

t=0.0
i=0.0
q=0.0
```

```python
tf=30.0 # final value of t
dt= 0.01 #step length
```

```python
tt=[]
ii=[]
qq=[]
```

```python
#defining Function
def f(Q,t): # Q=(q,i)
    q,i=Q
    f1=i
    f2=(e-i*r-q/c)/l
    return np.array([f1,f2])
def update(Q,t):
    k1=dt*f(Q,t) # dQ at initial point
    k2=dt*f(Q+0.5*k1,t+0.5*dt) 
    k3=dt*f(Q+0.5*k2,t+0.5*dt)
    k4=dt*f(Q+k3,t+dt)
    return Q+1/6.0*(k1+2*k2+2*k3+k4),t+dt

Q=np.array([q,i])
while t<=tf:
    tt.append(t)
    qq.append(Q[0])
    ii.append(Q[1])
    Q,t=update(Q,t)    
   
    
```

Plotting The current and Charge curve against time. 

To plot the behaviour of current and charge with respect to different cases, please open the locked variables by putting a "hashtag"/"#" before the code and change the plot code from RAW NBConvert to executable code type for respective case.

#Over Damped Case

```python
plt.subplot(2,1,1)
plt.plot(tt,ii,label="Current")
plt.xlabel("time")
plt.ylabel("Current")
plt.legend(loc='best')

plt.subplot(2,1,2)
plt.plot(tt,qq,label="Charge")
plt.xlabel("time")
plt.ylabel("Charge")
plt.legend(loc='best')
plt.suptitle("Solution of LCR circuit by RK4 method:Overdamped")
plt.savefig("Overdamped-numpy.png")
```
